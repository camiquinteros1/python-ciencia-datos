***APLICACIÓN DE PROBABILIDAD***

Se Realiza el lanzamiento de un dado EQUILIBRADO JUSTO, de manera reiterada con las mismas condiciones, y se anotan sus resultados= CARA O CRUZ.

Se le pide al usuario que defina la cantidad N de lanzamientos y se calcula su frecuencia relativa= 'PROBABILIDAD' de obtener cara o cruz. Se obtiene los siguientes resultados:

```python
import numpy as np
n=int(input('ingrese la cantidad de simulación o lanzamientos: '))

#========= FORMA 1 DE SIMULAR LANZAMIENTO DE UN DADO============
dado=[]
cara=0
cruz=0
for i in range(n):
  dado.append(np.random.rand(1))
  dado1=np.array(dado)
  if dado1[i]<0.5:
    #print('salio cara')
    #print(dado[i])
    cara +=1
  else:
    #print('salio cruz')
    #print(dado[i])
    cruz +=1
```

```python
#  mostrar la probailidad frecuencista de el lanzamiento de un dado
prom_cara=cara/n
prom_cruz=cruz/n
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CARA es:', prom_cara)
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CRUZ es:', prom_cruz)
```

```python
#========== FORMA COMPACTA DE SIMULAR USANDO NUMPY===========
m=int(input('ingrese la cantidad de simulación o lanzamientos: '))
lanzamientos = np.random.randint(0, 2, m)

# Calcular las probabilidades

probabilidad_cara = np.sum(lanzamientos ==0 ) / len(lanzamientos)
probabilidad_sello = np.sum(lanzamientos== 1) / len(lanzamientos)

print(f"Probabilidad de cara: {probabilidad_cara:.2f}")
print(f"Probabilidad de sello: {probabilidad_sello:.2f}")
```

```python
import matplotlib.pyplot as plt
x=np.array(range(2))
y=np.array([probabilidad_cara, probabilidad_sello])

plt.bar(x,y)
plt.xlabel('Opción')
plt.ylabel('Frecuencia relativa')
plt.title('Histograma del lanzamiento de una moneda equilibrada')
plt.show()
```

##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Lanzamiento de un dado
l=int(input('ingrese la cantidad de simulación o lanzamientos: '))
lanzamientos = np.random.randint(0, 7, l)

# Probabilidades

probabilidad_num2 = np.sum( lanzamientos == 2) / len(lanzamientos)
probabilidad_num3 = np.sum( lanzamientos == 3) / len(lanzamientos)
probabilidad_num5 = np.sum( lanzamientos == 5) / len(lanzamientos)

print(f"Probabilidad del numero 2: {probabilidad_num2:.3f}")
print(f"Probabilidad del numero 3: {probabilidad_num3:.3f}")
print(f"Probabilidad del numero 5: {probabilidad_num5:.3f}" )

# Histograma
x=np.array(range(3))
y=np.array([probabilidad_num2, probabilidad_num3, probabilidad_num5])

plt.bar(x,y)
plt.title('Histograma del lanzamiento de las probabilidades')
plt.show()
```

Solicitar al usuario que ingrese la cantidad N de datos a cargar. Luego generar de manera aleatoria los siguientes datos:

* SEXO= 'MASCULINO', 'FEMENINO','NO ESPECIFICA'
* EDAD= entre 28 y 60 años
* ALTURA= entre un rango de 165.0 mts hasta 210 mts
* PESO= entre un rango de 50  y 110 kg
* HORAS DE DESCANSO diario= un número decimal
* HORAS DE TRABAJO/ ESTUDIO diario= un número decimal
* HORAS DE EJERCICIOS diario= un número decimal

REalizar las siguientes indicaciones y extraer concluisiones sobre la información y la pooblación muestreada.

1.   Generar un array de dimensión 2 (matriz), con todos los datos ordenados, permitiendo visualizarlos como tabla.
2.   Suponiendo que la distribución de probabilidad de los datos muestrados de las N personas, se distribuyen en forma NORMAL. Encontrar la media en los datos de altura, peso, y horas de descanso, trabajo y ejercicios.
3.  Graficar la muestra o distribución de los datos  de pesos , y alturas en graficas separadas. Indicar el grado de variabilidad de sus datos
4.  Plotear de manera conjunta y encontrar si presentan correlación los datos de peso vs altura,
5. Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Solicitar al usuario la cantidad de datos a cargar
N = int(input("Ingrese la cantidad de datos a cargar: "))

# Generar datos aleatorios
np.random.seed(42)  # Para reproducibilidad
sexos = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], size=N)
edades = np.random.randint(28, 61, size=N)
alturas = np.random.uniform(165.0, 210.0, size=N)
pesos = np.random.uniform(50.0, 110.0, size=N)
horas_descanso = np.random.uniform(4.0, 9.0, size=N)
horas_trabajo_estudio = np.random.uniform(2.0, 12.0, size=N)
horas_ejercicio = np.random.uniform(0.5, 2.5, size=N)

# Tabla de todos los datos
data = {
    'Sexo': sexos,
    'Edad': edades,
    'Altura': alturas,
    'Peso': pesos,
    'Horas de Descanso': horas_descanso,
    'Horas de Trabajo/Estudio': horas_trabajo_estudio,
    'Horas de Ejercicio': horas_ejercicio
}
df = pd.DataFrame(data)
print(df)


# media de altura, peso y horas

print("Media de los datos (2)")
media_altura = np.mean(alturas)
media_peso = np.mean(pesos)
media_descanso = np.mean(horas_descanso)
media_trabajo_estudio = np.mean(horas_trabajo_estudio)
media_ejercicio = np.mean(horas_ejercicio)

print("\nMedias:")
print(f"Altura media: {media_altura:.2f} m")
print(f"Peso medio: {media_peso:.2f} kg")
print(f"Horas de Descanso media: {media_descanso:.2f} horas")
print(f"Horas de Trabajo/Estudio media: {media_trabajo_estudio:.2f} horas")
print(f"Horas de Ejercicio media: {media_ejercicio:.2f} horas")

# Grafico de los datos de pesos y alturas

print("Grafico de datos de pesos y alturas (3) ")

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(pesos, bins=20, color='blue', alpha=0.7)
plt.title('Distribución de Pesos')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
plt.subplot(1, 2, 2)
plt.hist(alturas, bins=20, color='green', alpha=0.7)
plt.title('Distribución de Alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# Grafico de las horas de descanso, trabajo/estudio y ejercicio
print("Grafico de horas de horas (5)")
plt.plot(horas_descanso, label='Descanso')
plt.plot(horas_trabajo_estudio, label='Trabajo/Estudio')
plt.plot(horas_ejercicio, label='Ejercicio')
plt.title('Horas de Descanso, Trabajo/Estudio y Ejercicio')
plt.xlabel('Personas')
plt.ylabel('Horas')
plt.legend()
plt.show()
```
